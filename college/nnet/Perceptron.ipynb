## Using Perceptrons

Perceptron were the first generation of neural networks.

In statistical pattern recognition, there's a standard way to recognize patterns:

* Convert the raw input vector into a set (or vector) of feature activations.
* Learn how to weight each of the feature activations to get a single scalar quantity, which represents our confidence that the data meets the pattern we want it to meet
* If the confidence is above some threshold, decide that the input vector is an example of the target class.

Explained individually:

### Convert raw input into feature activations

To convert raw input to a vector of feature activations, we make simple rules using common sense.  That part of the system does not learn.  The conversion tries to turn the data into something that it is easy to learn _on_.  We look at the problem, see what the features should be, if they don't work, we try some different ones, and eventually, we get a set of features that allow us to solve the problem by using a subsequent learning stage.

### Learn how to weight each feature activation to get a single scalar value

What we _learn_ is how to weight each of the feature activations to get a single quantity.  The weights represent how much evidence the feature gives you that the current evidence is an example of the kind of pattern you want to recognize.  When we add up all of the weighted features, we get the "total evidence" in favor of the hypothesis that this is the kind of pattern we want to recognize.

### If the quantity is above some threshold, decide that the input vector is an example of the target class

If the total sum of all of the weighted features is above some threshold (which we choose), we can decide that the input is an example of the kind of thing that we want to recognize.  

A _perceptron_ is a particular example of a statistical pattern recognition system.  There are many different kinds of perceptron, but the standard kind consists of some inputs, which are converted into feature activities.  That stage of the system does not learn.  Once you have the features, you then learn some weights, so that you can take the feature variables times the weights, and you decide whether or not the input is an example of the class you're interested in, by seeing if that sum of feature activities times learned weights is greater than the input threshold.
