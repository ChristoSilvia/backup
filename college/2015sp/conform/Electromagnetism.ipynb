```julia
using PyPlot
```

```julia
a = 1.0
```

```julia
σ = 1.0
```

```julia
vacuum_permitivity = 1.0
```

```julia
function rod_potential(r)
    (a ./ r) .* σ .* vacuum_permitivity
end
```

```julia
function plate(z)
    z + 1.0./z
end
```

```julia
function plate_inverse(z)
    #positive branch
    z./2.0 + sign(real(z)).*sqrt(z.^2 ./ 4.0 - 1.0)
end
```

```julia
function plate_potential(z)
    rod_potential(abs(plate_inverse(z)))
end
```

```julia
Z = linspace(1,5) * exp(im * linspace(0,2*pi))';
```

```julia
plot(real(Z'),imag(Z'));
```

```julia
surf(real(Z), imag(Z), rod_potential(abs(Z)))
```

```julia
plot(real(plate(Z')), imag(plate(Z')))
axis([-3,3,-1,1]);
```

```julia
plot(real(plate_inverse(plate(Z'))), imag(plate_inverse(plate(Z'))));
```
