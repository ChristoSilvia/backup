```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
```

```python
a = 0
b = 0.9
n = 100
```

```python
initial_condition = 1.0
```

```python
def first_order(f,x):
    return f**2
```

```python
X = np.linspace(a,b,n)
```

```python
Y = odeint(first_order, initial_condition, X)
plt.figure(figsize=(14,8))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("First Order Differential Equation Plot")
plt.grid(True)
plt.plot(X,Y,label="f(x)")
plt.legend()
```

```python
def second_order(X,t):
    return np.array([X[1], -5*9.8])
```

```python
initial_slope = 1.0
```

```python
Z = odeint(second_order, np.array([initial_condition, initial_slope]), X)
plt.figure(figsize=(14,8))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Second Order Differential Equation Plot")
plt.plot(X,Z.T[0], label='f(x)')
plt.plot(X,Z.T[1], label='df/dx(x)')
plt.grid(True)
plt.legend()
```
