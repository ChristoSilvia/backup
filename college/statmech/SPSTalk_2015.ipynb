##Some quick thermodynamics
* We often study thermodynamics in the context of heat engines. Why is this? Thermodynamics is the study of turning heat into something useful, work! So the simplest way to do this is to have two heat reservoirs, one 'hot' at temperature $T_h$ and one 'cold' at temperature $T_c$, and some "process" in between. Needless to say, $T_h>T_c$.

* So a heat engine takes heat from the hot reservoir, call that amount of heat energy $Q_h$, and we get work out, and then the cycle dumps the energy not turned into work into the cold reservoir.

    * We can define the efficiency as $e = \frac{W}{Q_h}$, so that this system is maximally efficient when all the heat we extract is converted into work.
    
    * It turns out that the __best we can ever do__ is $e = 1 - \frac{T_c}{T_h}$
    
    * You can see this will never be unity, because we would need either $T_c=0$ or $T_h\rightarrow\infty$

* __Why is there some maximum efficiency?__

    * Another condition for a maximally efficient heat engine is that $\frac{Q_h}{T_h} + \frac{Q_c}{T_c} = 0$
    
    * We can call this $Q/T$ quantity $S$, and since $Q$ is by definition a change in energy
    
        * $\frac{Q_h}{T_h} + \frac{Q_c}{T_c} = \Delta S_h + \Delta S_c = \Delta S_{Total}=0$
         
    * It turns out if we run a process that isn't maximally efficient, this strange quantity $\Delta S >0$, so that we can characterize all processes as being either:
        
        * __Reversible__ so that $\Delta S =0$ or
        * __Irreversible__ wherein $\Delta S >0$

    * This quantity $S$ is called the __entropy__ of a system, and it is the reason we can't be maximally efficient.
    
    * This is the so-called "arrow of time", i.e. properties of entropy (the 2nd law of thermodynamics) are the only known laws of physics that distinguish between "future" and "past"

## Statistical Mechanics

* The previous discussion is great, but with it we can only understand large-scale things like total pressure $P$, total volume $V$, and temperature $T$ (whatever that is!)

* Given a container of gas with $P$, $V$, and $T$ it's clear that even though these things are constant, there is a lot of motion among the constituent atoms. 

* __How do we understand these microscopic dynamics?__

* Consider a silly model. Say our world is a string of 0's and 1's, and we have $N$ of them, but we can only 'see' the sum of all the 1's and 0's, i.e., the number of 1's $N_1$. This is just like a box of gas, where we can only see the pressure for instance.

    * The world: 01001101111100111000011111111000101001010100...

* How many ways can our world be arranged with the same sum? Lots of ways, with some factorials for sure. We will call this number of ways we can arrange the 1's and 0's to be $W$. The name of this quantity is the __multiplicity__. 

    * We can calculate it by considering how we can make a 'world' string like the one above. Say we start with $N$ 0's, then we can ask, for a given number of 1's $N_1$, how many ways can we choose 0's to change to 1's? This is a standard combinatoric problem of choosing $N_1$ objects from $N$ without putting them back after choosing so that $W = \binom{N}{N_1} = \frac{N!}{N_1!(N-N_1)!}$

* It turns out the entropy of this 'world' is $S=k_B\log W$, where $k_B$ is the boltzmann constant

* This can help us understand the 'arrow of time'-ness of entropy: If all states are equal and we randomly see any state at a time, the 'sum' ($N_1$) with the highest $W$ is the one we are most likely to find! If we break a vase, we never see the vase spontaneously un-break because there are __many more ways for something to be broken than not broken__.

## Probability and Information

* We have been talking about states and probabilities without being precise. Empirically, probabilities are just histograms of counting how many times we see things. If we look at our binary world above, and randomly see some configuration (I'm going to call it $s^{(0)}$, where an individual element is $s^{(0)}_i = 0$ or $1$), we need to make many observations to understand what the world 'typically' looks like.

* Say we make $M$ observations of the world, denoted by $s^{(0)}, s^{(1)}, ..., s^{(M-1)}$. There are some copies in here, they are not all unique. So we actually see $L<M$ unique states, and for each unique $s$, we count them up and divide by $M$. We now have an estimate of the proportion of each unique state $s^l$ states in the world $p_l$. This probability $p_l$ (which is between 0 and 1) has the property that if we add up all of them $\sum_l p_l = 1$ so that if we saw everything, there would be nothing missing!

* Entropy can still be defined in terms of these probabilities, and is much more general than our world of 0'1 and 1's or even thermodynamics!

    * For a probability distribution $p_l$ we define Entropy to be $S= -k_B \sum_l p_l \log p_l$
    
    * We can see that this is consistent with our assertions above, by going to the limit of all states being equally likely. In this case $p_l = 1/L$ where $L$ is the number of unique states. Then $S = -k_B \sum_l \frac{1}{L} \log \frac{1}{L} = -k_B \sum_l \frac{1}{L} (\log 1 - \log L) = k_B \sum_l \frac{1}{L} \log L = k_B \log L$. This is the same formula as above, because $L$ is the __multiplicity__. 
    
    * $S= -k_B \sum_l p_l \log p_l$: So this is a more general definition of entropy, as it applies to nonuniform probabilities
    
* __How do we interpret $S$ in this context?__

    * When probabilities are uniform it is just the logarithm of the number of possibilites, what about in general?
   
    * An important property of this formula for entropy is that __it is maximal for uniform probability distributions__, i.e. when all states are equally likely, a single sample or view of the world will tell you the least about the world!
       
    * It turns out to be understood as __information content__:
    
        * Consider our world as a string with a single element ($N=1$). To understand this world is to know the probability of 0 and 1. Each time we look at it, it's either 1 or 0. Instead of writing down 1's and 0's, we can also just track when the 1's switch to zeros and vice versa. If the world is uniform, i.e. $P(s=1)=P(s=0)=0.5$, the entropy is $S=k_B\log2$ and on average we have to write down that the world switched each time we observe it. Contrast this with a world that is almost always 1, wherein $S\approx0$, we very rarely have to note that it switches to a 0 and back to 1.
        
        * Then entropy is the average amount of information we learn about the world from a single observation

    * It almost sounds like two opposite things are true. If we have maximal entropy, the average information content of a single observation is large. When we make one observation then, predicting subsequent observations is difficult because each one has so much information. __Therefore we are maximally uncertain about subsequent observations when entropy is maximal.__
    

## Maximum Entropy as Occam's Razor

* Occam's Razor: When trying to explain something, among all competing hypotheses, the one with the fewest assumptions should be selected.

* __Question: In statistical mechanics, how should we assign probabilities to states?__

* If we have $L$ states denoted by $s^l$, and each has an energy $E_l$, what is most important about this data? We make granular observations, but the only well-defined quantity is the average energy $E_{av} = \langle E \rangle = \sum_l E_l p_l$ where $p_l$ is the probability of state $s^l$.

* If we make a probability distribution which maximizes the entropy, while restricting $E_{av} = \langle E \rangle$, this is in some sense the simplest possible model which can represent our data. In the context of thermodynamics it means we want the model to have the highest multiplicity, and in the context of information it means that we should be maximally uncertain about future observations based on our current data.

* Maximum entropy probability distributions are a kind of quantitative expression of Occam's Razor!

* So what is the max-ent distribution for the above problem?

    * The problem restated: find the $p_l$ which maximizes $S=-k_b\sum_l p_l \log p_l$ such that $\sum_l E_l p_l = E_{av}$
    * This is a constrained optimization problem which can be solved with Lagrange multipliers!
    * The solution is $p_l = e^{-\beta E_l}/\mathcal{Z}$ where $\mathcal{Z} = \sum_le^{-\beta E_l}$, the __Boltzmann distribution which is the fundamental result of statistical physics!__
    
* It turns out this procedure is quite general. Say we have our binary world, with observataion $s^{(0)}, s^{(1)}, ..., s^{(M-1)}$, but we think that the most important property is the correlation between any two spots. What I mean by correlation is, if we know that one of these digits is 1, what is the probability that another digit will be 1? We denote and compute it this way: $\langle s_i s_j \rangle = \frac{1}{M} \sum_l s_i^{(l)} s_j^{(l)} p_l$. If we compute the maximum entropy distribution constrained to have these correlations, the maximum entropy distribution is:

    * $p(s) = e^{-\beta \mathcal{H}(s}/\mathcal{Z}$ where $\mathcal{H} = -\sum J_{ij} s_i s_j$
    
    * The max-ent distribution for binary data where correlations matter is the famous Ising model of magnetism!

##How hard is it to study these models numerically?

* We can always compute the energy of a given state of our model, i.e. for all spins $s_i = +1$ we have $\mathcal{H} = -J N $ where $N$ is the number of spins or the 'length' of our system. We often set $J=1$ for convenience, which is essentially using a special unit of temperature, where one unit of temperature is equal to the interaction strength $J$. We also often set $k_B=1$, so that our model has really only one parameter, $T$.

* Not all the states of our model are equally likely. Only in the limit of very high temperature are all states equally likely, and this is the most boring limit because things are perfectly random!

* So some states are more likely than others, and we know that the probability of any given state is $P(s) = e^{-\mathcal{H}/T}/\mathcal{Z}$. Can we compute these probabilties? We can certainly compute the numerator, but the denominator has $2^N$ terms in it!

* If we had $N=50$ spins in our model (a pathetically small system!), then $2^N \approx 1.2\times 10^{15}$. This would take a single modern CPU __1 week__ of constant effort _just to list out every term in $\mathcal{Z}$_, nevermind compute anything interesting!

* This is very clearly a deadend direction for almost every model you could think of, so how do we proceed?

# Monte Carlo Sampling

* Sampling is very useful. For example, say we didn't know what a gaussian looked like. We could draw samples from the probability distribution of a gaussian to try to see it. The more we draw, the more it looks like a gaussian.

```python
from numpy.random import randn
fig, axs = plt.subplots(1,4,figsize=(15,4))
bin_numbers = [int(10**n) for n in range(2,6)]
for bn, ax in zip(bin_numbers, axs.flat):
    ax.hist(randn(bn), bins=100, normed=True, alpha=1);
    ax.set_title('{} Samples'.format(bn))
plt.show()
```

* Often we want to compute something of interest, like an expectation value: $E[f(x)]=\int f(x)p(x)dx $ But, doing this calculation can be complicated or even analytically intractable.

* If that's the case, the first thing we could try would be to discretize the integral. Recall that discrete approximations converge to the integral as the discrete pieces get smaller $\Rightarrow E[f(x)] \approx \sum f(x)p(x)$ 

* Oftentimes, as mentioned above, it still is infeasible to generate all of these pieces. But we can generate a finite set of  them. And, if we do it right, we can get an accurate answer.  

* The way to proceed, then, is to generate samples from the probability distribution. This can be done using Markov Chain Monte Carlo.

Gibbs Sampling
---------------

* Gibbs sampling is a really nice way to sample from joint probability distributions where all of the probabilities can be written in conditional form.

* The algorithm:
    * $x^{(0)} = (x_1^{(0)},x_2^{(0)},...x_k^{(0)})$
    * for t=1 to T do:
        * for i=1 to k:
            * $x_i^{(t+1)}\sim P(X_i|x_1^{(t+1)},...x_{i-1}^{(t+1)},x_{i+1}^{(t+1)}, ...x_k^{(t+1)})$
        * end for
    * end for

* For the Ising model, the probability of a state of the whole system is $P(s) = e^{-\mathcal{H}/T}/\mathcal{Z}$ where each site $s_i \in \{+1, -1\}$. But, we also have a way to write the probability that a given site is +1 given it's neighbors which is much simpler.

* This conditional version is $P(s_i=+1|neighbors) = e^{- H_i/T }/\mathcal{Z}$ where $H_i = -J \sum_{neighbors} s_i s_j$ and $s_i = +1$.

* Since $s_i = \pm 1$ and the neighbors are fixed, $\mathcal{Z}$ is easy now! The sum is only over $s_i$ which can only take   one of two values ${+1,-1}$.

* We can write the Gibbs sampling algorithm in words for the Ising model:
    * Pick a state of 1's and -1's.
    * For each site determine the probability, p, that it is a 1 based on it's neighbors.
    * Set it to 1 with that probability (and to -1 with probability $p'=1-p$). 
    * Repeat for as many time steps as you need to generate a fair sample of states.

* You may also update multiple sites at once as long as they are not directly conditionally dependent on eachother.  

* What if you don't have conditional probabilities?

Metropolis Hastings
--------------------

* There are a lot of other types of sampling you can try: Metropolis-Hastings, Slice Sampling, Hamiltonian Monte Carlo, Parallel Tempering, ... The next simplest but multipurpose sampling algorithm is Metropolis Hastings. You can think of it as a random walk in parameter space biased toward more probable regions.

* The work horse of MH is detailed balance. "Detailed balance" is a sufficient (but not necessary) condition that ensures the distribution is stationary (i.e. it's possible to converge to it...). Mathematically, it is written: $ P(x)T(x\rightarrow x')=P(x')T(x'\rightarrow x) $. 

* $P(x)$ is the probability of being at a point x. $T(x\rightarrow x')$ is the probability of transitioning from $x$ to $x'$.

* The transition probability is given by: $ T(x\rightarrow x')=\Gamma(x\rightarrow x')A(x\rightarrow x') $

* Simply plugging this in gives us a formula for the ratio of the acceptance probabilities: $ \frac{A(x\rightarrow x')}{A(x'\rightarrow x)}=\frac{P(x')\Gamma(x'\rightarrow x)}{P(x)\Gamma(x\rightarrow x')} $ 

* The Metropolis Hastings algorithm defines: $ A(x\rightarrow x')=min(1, \frac{P(x')\Gamma(x'\rightarrow x)}{P(x)\Gamma(x\rightarrow x')}) $ 

* The algorithm:
    * Propose a step (Typically a random gaussian step from the current point)
    * Compute the acceptance probability
    * Draw a random number between 0 and 1 to decide whether to accept/reject
    * If accept add the new point to the chain, if reject add the old point again
    * Repeat
    
* Most people shoot for an acceptance rate of 0.2 to 0.5 although the optimal will vary with the problem.
    
Notes of Caveat:
------------------

* It takes time to converge. You probably won't know how much time until it occurs. (Even then you should be wary of being fooled!! For example, the distribution could be multi-modal.)

* If steps are highly correlated, you have to take a lot of them to gather an independent sample.

Demo
----

* http://physics.weber.edu/schroeder/software/demos/IsingModel.html

```python
def fixed_step_metro_hastings(f, iterations, step_size, init_location):
    points = []
    locations = []
    
    x = init_location
    old = f(init_location)
    
    for i in range(0,iterations):
        rn = np.random.randn()
        new = f(x + rn*step_size)
        if np.random.uniform() < np.abs(new/old):
            x = x + rn*step_size
            old = new
            points.append(x)
            locations.append(new)
    return points, locations
```

```python
X, Y = fixed_step_metro_hastings(lambda x: x**2*np.exp(-x**2), 1000, 0.5, 0.0)
```

##Notes on the one dimensional Ising Model
* Energy function is $\mathcal{H} = -\sum_i J s_i s_{i+1}$ where $s_i \in \{+1, -1\}$

    * Notice that if $J>0$ the lowest energy configuration of $s$ is all +1 or -1 (all pointing in same direction)
    
    * If $J < 0$ then the spins like to alternate +1 -1 +1 -1... etc

* Probability of some configuration is $P(s) \propto e^{-\frac{\mathcal{H}}{T}}$

    * The spin configuration with the lowest energy has the highest probability

    * Normalization of $P(s)$ is called the partition function $\mathcal{Z} = \sum_{\{s\}}e^{-\frac{\mathcal{H}}{T}}$
    
        * This sum $\sum_{\{s\}}$ is a sum over all __possible configurations__ of spins $s_i$, and is the most difficult part of these calculations
        
        * If there are $N$ spins, then there are $2^N$ possible configurations!

* All physical quantities can be computed from the _free energy_ $\mathcal{F}$

    * $\mathcal{F} = -\frac{1}{k_B T}\log{\mathcal{Z}}$
    
* We consider this model 'solved' because we can write down $\mathcal{F}$ as a function of $k_B$, $T$, and $J$ alone

    * If you're curious, $\mathcal{F} = -\frac{1}{k_B T} \log \left( \text{cosh}(\frac{J}{T}) \right)$


```python
F = lambda T: -np.log(np.cosh(1./T))/T
Ts = np.arange(0.1,1., 0.01)
plt.plot(Ts, F(Ts));
plt.ylabel('Free energy');
plt.xlabel('Temperature');
plt.title('Free energy of 1D Ising Model')
```

* $\mathcal{F}$ is a smooth function for all $T$, only diverges (goes to $-\infty$) as $T\rightarrow 0$

    * This means that the 1D Ising model has _no phase transition_
    * i.e. It never 'freezes' at $T>0$, where freezing means that all the $s_i$ align
    
* Ising was discouraged by this result, as he was studying this model to understand ferromagnetism. In real magnets, at some temperature (the Curie temperature), the material can support a spontaneous magnetization. This 1D model never freezes, and so cannot explain ferromagnetism.

* As it turns out, the 2D Ising model _does_ have a nonzero phase transition, and shares many of the same properties as a real ferromagnet. This is the most famous example of a _Universality Class_, where ostensibly different physics systems share some quantity. The shared quantities turn out to be exponents of quantities like the correlation length, heat capacity, etc.

* The 2D Ising model does have a solution, but it is __much__ more difficult to achieve than the 1D. The 3D Ising model has no known solution.

* There are very few analytically tractable models out there, so how can we proceed with more complicated systems?

#Sampling a 1D Ising model

* We'll consider a ring of $N$ 'spins', where $s_N = s_0$ (The ring wraps around).

* __How do we compute our transition probabilities for our Markov chain?__

    * Let's propose a step as follows: Given some state $s$, we will try flipping one of the spins $s_i\rightarrow -s_i$

    * Question: Given some state $s$, what is the ratio of probabilities $P(s, s_i\rightarrow-s_i)/P(s)$. That is, what is the probability of our system being in a state with $-s_i$ compared to the probability of the same state except with $s_i$?
    
    * We know $P(s) \propto \exp -\beta \sum_i s_i s_{i+1}$. We are ignoring the normalization because when we take the ratio of probabilities it cancels. Notice that if we switch one spin, only two of the terms change, and they turn into opposites of each other.
    
    * This ratio of probabilities is then $P(s, s_i\rightarrow-s_i)/P(s) = \exp({2\beta Js_i(s_{i-1} + s_{i+1})})$
    
    * The algorithm is then, given a list of spins $s$:
    
        for i in [0,N-1]:
        1. Compute $P = \exp({2\beta Js_i(s_{i-1} + s_{i+1})})$
        2. Draw a uniform random number $r$ from (0,1]
        3. If r<P, set $s_i = -s_i$
        
* Goals for the reader:

    1. Write a function which, given a spin list $s$, an index $i$, and a temperature $T$, computes the ratio of probabilities of site $i$ flipping versus not flipping.
    
    2. Write two for loops, that is, a for loop within another for loop. The outer for loop will run a specified number of times, call it N_sweeps. The inner for loop will go over the sites, trying to flip each spin according to the algorithm above. Then the spins array should be copied after each inner loop.
    
    3. For a small system of $N=8$, compare the probabilities generated by your sampling code to the analytic probabilities.
    
    4. Show that at $T=0.1$, the correlations between spins still exponentially decreases with distance, so that there is no phase transition.
    

```python
def p_switch(s,i,T):
    """
    Computes ratio of probabilitis of P(si->-si)P(si)
    s : list of spins
    i : integer denoting spin
    T : temperature
    """
    return np.exp(-2*s[i]*(s[(i-1)%N]+s[(i+1)%N])/T)
#Note: (i-1)%N always gets pulled back into the range (0 to N-1), so 
#       This is how we implement periodicity
```

Sample N_sweep states of N-length system at temperature T

```python
N = 4 #Number of spins
T = 1.0 #Temperature
N_sweeps = 10000 #Number of Monte Carlo Sweeps

initial_s = np.ones(N)#2*(np.random.rand(N)>0.5)-1 #Random +1,-1
samples = []
for i in range(N_sweeps):
    for i in range(N):
        r = np.random.rand()
        P = p_switch(initial_s, i, T)
        if r < P:
            initial_s[i] = -1*initial_s[i]
    samples += [initial_s.copy()]
            
```

What follows is code that computes the probabilities exactly. Don't worry about this part, it's just to compare to what your sampling procedure generates.

```python
from itertools import product
all_possible_states = np.array(map(list, list(product([-1,1],repeat=4))))
def Hamiltonian(s):
    return -np.sum(np.roll(s,1)*s)
def proportional_to_prob(s, T):
    return np.exp(-Hamiltonian(s)/T)
def Partition_function(T):
    return np.sum(map(lambda s: proportional_to_prob(s, T),
                      all_possible_states))
Z = Partition_function(T)
probabilities = np.array(map(lambda s: proportional_to_prob(s,T), 
                    all_possible_states))/Z
```

```python
def spinsToInt(spins):
    """
    Converts a spin array into an integer so we can count individual states
    """
    binary = ''.join(map(str,(map(int,1*(spins>0).ravel()))))
    return int(binary,2)
```

```python
spinintegers = map(spinsToInt, all_possible_states)
sampledspinintegers = map(spinsToInt, samples)
from collections import Counter
count = Counter(sampledspinintegers)
plt.plot(spinintegers, probabilities, label='Exact probabilities')
sortkeys = np.argsort(count.keys())
plt.plot(np.array(count.keys())[sortkeys], 
         np.array(count.values())[sortkeys]/float(len(samples)), 
         label='Sampled probabilities')
plt.legend()
```
