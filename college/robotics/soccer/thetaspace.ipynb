```python
def min_distance(a, theta,l,x_desired, y_desired):
    return np.abs((a-x_desired)*np.cos(theta) - (y_desired - l/2.0)*np.sin(theta))
```

```python
na = 1000
ntheta = 1000
a = np.linspace(-w/2, w/2, na)
theta = np.linspace(-np.pi/2, np.pi/2, ntheta)
A, Theta = np.meshgrid(a, theta)
```

```python
mask = np.zeros((na, ntheta), dtype="bool")
n_bounce_limit = 3
for n in xrange(-n_bounce_limit, n_bounce_limit+1):
    mask |= (Theta > np.arctan2(n*w-(wg/2.0 + A), l/2.0))&(Theta < np.arctan2(n*w + wg/2.0 - A, l/2.0))
```

```python
distances = 0.8*(~mask) + mask*min_distance(A, Theta, l, 0.1, 0.3)
```

```python
plt.figure(figsize=(10,7))
for n in xrange(-nlim, nlim+1):
    plt.plot(a, np.arctan2(n*w + wg/2.0 - a, l/2.0), color="red")
    plt.plot(a, np.arctan2(n*w -(wg/2.0 + a), l/2.0), color="blue")
plt.pcolor(A, Theta, distances)
```
