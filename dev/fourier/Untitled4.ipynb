```python
%pylab inline

import pylab as pl
import numpy as np
```

```python
from sklearn.datasets import fetch_lfw_people

lfw_people = fetch_lfw_people(min_faces_per_person=69, resize=0.4)
```

```python
X = lfw_people.data
y = lfw_people.target
names = lfw_people.target_names

n_samples, n_features = X.shape
_, h, w = lfw_people.images.shape
n_classes = len(names)

print("n_samples: {}".format(n_samples))
print("n_features: {}".format(n_features))
print("n_classes: {}".format(n_classes))
```

```python
def plot_gallery(images, titles, h, w, n_row=3, n_col=6):
  pl.figure(figsize=(1.7 * ncol, 2.3 * nrow))
  pl.subplots_adjust(bottom=0, left=0.01, right=.99, top=.90, hspace=.35)
  for i in range(nrow * ncol):
      pl.subplot(nrow, ncol, i+1)
      pl.imshow(images[i].reshape((height,width)), cmap=pl.cm.gray)
      pl.title(titles[i], size=12)
      pl.xticks(())
      pl.yticks(())
        
plot_gallery(X, names[y], h, w)
```

```python
def repartition_bar(labels, names_list):
  pl.figure(figsize=(14,3))
  
  y_unique = np.unique(labels)
  counts = [(y == i).sum() for i in y_unique]
  
  pl.xticks(y_unique, names_list[y_unique])
  locs, labels = pl.xticks()
  pl.setp(labels, rotation=45, size=20) 
  _ = pl.bar(y_unique,counts)
    
repartition_bar(y,names)
```

We can split the data in a development set and final evaluation set

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

To train a model we will first reduce the dimensionality of the original picture to a 150-dimensional space of eigenfaces.

```python
from sklearn.decomposition import RandomizedPCA

def do_pca_model(dataset, n_components):
  print "Extracting {0} eigenfaces from {1} faces".format(150,dataset.shape[0])
  
  pca = RandomizedPCA(n_components=150, whiten=True) 
  
  %time pca.fit(dataset)
  
  return pca.components_.reshape((150, h, w))

eigenfaces = do_pca_model(X_train, 150)
```

```python
eigenface_titles = ["eigenface {0}".format(i) for i in range(eigenfaces.shape[0])]
plot_gallery(eigenfaces, eigenface_titles, h, w)
```

```python
X_train_pca = pca.transform(X_train)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', class_weight='auto')

svm
```

```python
from sklearn.cross_validation import StratifiedShuffleSplit
from sklearn.cross_validation import cross_val_score

cv = StratifiedShuffleSplit(y_train, test_size=0.20, n_iter=3)

%time svm_cv_scores = cross_val_score(svm, X_train_pca, y_train, scoring='f1', n_jobs=2)
svm_cv_scores
```

```python
svm_cv_scores.mean(), svm_cv_scores.std()
```

```python
from sklearn.grid_search import GridSearchCV

param_grid = {
  'C': [1e3, 5e3, 1e4, 5e4, 1e5],
  'gamma': [0.0001,0.0005,0.001,0.005,0.01,0.1],
}

clf = GridSearchCV(svm, param_grid, scoring='f1', cv=cv, n_jobs=2)
```

```python
%time clf = clf.fit(X_train_pca, y_train)
print("Best extimator found by randomized hyper parameter search:")
print(clf.best_params_)
print("Best parameters validation csore: {:.3f}".format(clf.best_score))
```
