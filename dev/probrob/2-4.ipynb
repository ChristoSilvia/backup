```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,2000,400)
```

```python
xinit = 1000 #m
sinit = 30 #m

Prior = (2 * np.pi * sinit**2)**(-0.5) * np.exp(-0.5*(X - xinit)**2/sinit**2)
```

```python
zgps = 1100 #m
sgps = 10 #m

Measurement = (2 * np.pi * sgps**2)**(-0.5) * np.exp(-0.5*(X - zgps)**2/sgps**2)
```

```python
plt.plot(X,Prior)
```

```python
plt.plot(X,Measurement)
```

```python
plt.plot(X,Prior * Measurement)
```
