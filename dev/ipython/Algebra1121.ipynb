## Decomposing Vector Spaces and Modules into direct sums of things

Let $R$ be a PID. 
Let $M$ be a finitely generated module over $R$.
We're going to try to decompose $M$ into a direct sum of cyclic modules, (which are generated by one element).
Some examples of cyclic modules are $Z$ and $Z/m$ for $m$ an integer

In the fields case, let $V$ have dimension 1 over a field $F$.
Since $V$ is one dimensional, it is a cyclic $F$-module.  Consider the map $\phi:F \to V$ which takes $\phi(r) \to r v_0$.
This is an isomorphism so $F \approx V$

Now, think about $R$ a PID.  Recall that cyclic modules can be generated by one element.  For a module $M$ let $m_0$ be an element that generates it.

Consider:
$$
f:R \to M
$$
$$
f(r) \to r m_0
$$

This is injective, and to make it surjective, we take:
therefore $R/Ker(f) \approx M$.

### For Fields and Vector Spaces

Consider the exact sequence:
$$F^n - T \to F^m - f \to V \to 0$$
And let $A$ be a basis for $F^n$ and $B$ be a basis for $F^m$.
For the linear transformation $T$, write its matrix form: $[T]_{A,B}$.
Now that we know what the image of $T$ is, and it has rank at most $n$. Let $k \leq n$.
Since $F^m \approx F \oplus F \oplus \dots k \dots \oplus F$.

Then, $V \approx F \oplus F \oplus \dots n-k \dots \oplus F \oplus 0 \oplus \dots \oplus 0 \approx F^{m-k}$


### For Rings(PID's) and Modules

Let M be a free module over $m$ elements.  Consider the exact sequence:

$$R^m - S \to M \to 0$$

Note $Ker(S) \subseteq R^m$.  We can then construct a map $T: R^n \to R^m$ such that $Im(T) = Ker(S)$. This makes this sequence exact:

$$R^n \to^T R^m \to^S M \to 0$$


Then, write $T$ in matrix form: $[T]_{A,B}$.  After some row operations, we get bases $[T]_{A',B'}$ which gives something like:
$$\left( \begin{array}{ccc}
a_1 & \dots & 0\\
\vdots & \ddots & \vdots\\
0 & \dots & a_n \end{array} \right)$$
