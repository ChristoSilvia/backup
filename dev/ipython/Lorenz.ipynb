```julia
using PyPlot
using ODE;
```

```julia
function f(t, r)
    (x, y, z) = r
    
    dx_dt = sigma*(y - x)
    dy_dt = x*(rho - z) - y
    dz_dt = x*y - bet*z
    
    [dx_dt, dy_dt, dz_dt]
end;
```

```julia
dt = 0.001
tf = 100.0
t = [0:dt:tf]

r0 = [0.1, 0.0, 0.0]

sigma = 10.0
rho = 28.0
bet = 8.0/3.0;
```

```julia
function expon(t, x)
    -x
end
```

```julia
(t,y) = ode23(expon, 3.0, [0:0.01:1]);
```

```julia
y
```

```julia
plot(t,y)
```

```julia
(t,pos) = ode23(f, r0, t);
```

```julia
x = zeros(length(pos))
y = zeros(length(pos))
z = zeros(length(pos))
for i in 1:length(pos)
    x[i] = pos[i][1]
    y[i] = pos[i][2]
    z[i] = pos[i][3]
end
```

```julia
plot3D(y, x , z);
```
