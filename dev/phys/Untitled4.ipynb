```julia
using Gadfly
```

```julia
using Interact
```

```julia
@manipulate for phi = 0:pi/16.0:4.0*pi, f=[:sin => sin, :cos => cos]
    plot(theta -> f(theta + phi), 0, 25)
end
```

```julia
function lorentzian(x,a,sigma)
    a/(sigma + x.^2)
end
```

```julia
function gaussian(x,a,sigma)
    a*exp(-sigma*x.^2)
end
```

```julia
@manipulate for sigma=0.1:0.1:10, A=0.1:0.1:10, f = [:gaussian => gaussian, :lorentzian => lorentzian]
    plot(x -> f(x,A,sigma),-10,10)
end
```
