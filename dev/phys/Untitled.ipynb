```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PointTree1():
    def __init__(self, points):
        self.center = np.mean(points)
        if len(points) > 1:
            self.subtrees = [PointTree1(np.array(filter(lambda x: x < self.center, points))),
                             PointTree1(np.array(filter(lambda x: x > self.center, points)))]
        else:
            self.subtrees = []
            
    def __str__(self):
        if self.subtrees is []:
            return str(self.center)
        else:
            return "\n".join(str(self.center),str(self.subtrees))
```

```python
print PointTree1(np.random.randn(10))
```

```python
class PointTree4():
    def __init__(self, points):
        data_length, _ = points.shape
        self.center = np.mean(points,axis=0)
        
        subtree_accumulator = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
        for point in points:
            index = 0
            if point[0] > self.center[0]:
                index += 8
            if point[1] > self.center[1]:
                index += 4
            if point[2] > self.center[2]:
                index += 2
            if point[3] > self.center[3]:
                index += 1
            subtree_accumulator[index].append(point)
        
        self.subtrees = [ PointTree4(np.array(point_list)) if point_list is not [] else [] for point_list in subtree_accumulator ]
```
